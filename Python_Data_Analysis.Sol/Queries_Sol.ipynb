```python
import pandas as pd
import matplotlib.pyplot as plt
store=pd.read_csv('superstore.csv')
plt.style.use('ggplot')
stnd_class=store.loc[store['Ship Mode']=='Standard Class'].count()[0]
first_class=store.loc[store['Ship Mode']=='First Class'].count()[0]
secnd_class=store.loc[store['Ship Mode']=='Second Class'].count()[0]
same_day=store.loc[store['Ship Mode']=='Same Day'].count()[0]
explode=(.1,.1,0,.1)
plt.pie([stnd_class,first_class,secnd_class,same_day],labels=['Standard Class','First Class','Second Class','Same Day'],autopct='%.1f %%',pctdistance=0.4,explode=explode)
plt.show()


    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Clean the 'Discount' column
store['Discount'] = pd.to_numeric(store['Discount'].str.replace('%', ''), errors='coerce')

# Step 2: Calculate the average discount for each region
average_discount = store.groupby('Region')['Discount'].mean()

# Step 3: Plot the bar chart
average_discount.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount Across Regions')
plt.show()

# Step 4: Identify the region with the highest average discount
region_highest_discount = average_discount.idxmax()
print("Region with the highest average discount:", region_highest_discount)

```

# Insights

```python
average_profit = store.groupby('Region')['Profit'].mean()
plt.ylabel('Average Profits')
average_profit.plot(kind='bar')
```
