```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
df = pd.read_csv('loan.csv', low_memory=False)


```

```python
plt.style.use('ggplot')

grades = df['grade'].value_counts()

plt.pie(grades, labels=grades.index, autopct='%1.1f%%')

plt.title('Distribution of Grades')

plt.show()
```

```python

grade = df['grade']
income = df['annual_inc']

# Calculate the average income for each grade
avg_income = df.groupby('grade')['annual_inc'].mean().sort_index()


plt.bar(avg_income.index, avg_income.values)


plt.xlabel('Grade')
plt.ylabel('Average Income')
plt.title('Average Income vs. Grade')


plt.show()
```

```python

grade = df['grade']
installment = df['installment']

# Calculate the average installment for each grade
avg_installment = df.groupby('grade')['installment'].mean().sort_index()

sns.barplot(x=avg_installment.index, y=avg_installment.values)

plt.xlabel('Grade')
plt.ylabel('Average Installment')
plt.title('Average Installment vs. Grade')

plt.show()
```

```python
loan_grade_avg_amnt = df.groupby('grade')['loan_amnt'].mean()

plt.figure(figsize=(8,6))

plt.style.use('ggplot')
plt.bar(loan_grade_avg_amnt.index,loan_grade_avg_amnt.values)

plt.xlabel('Grade')
plt.ylabel('Average Loan Amount')
plt.title('Average Loan Amount by Grade')
plt.show()
```

```python

grade = df['grade']
int_rate = df['int_rate']
numeric_rates = [float(rate.strip('%')) for rate in int_rate]
df['int_rate'] = numeric_rates
avg_int_rate = df.groupby('grade')['int_rate'].mean().sort_index()
plt.bar(avg_int_rate.index, avg_int_rate.values)
plt.xlabel('Grade')
plt.ylabel('Average Interest Rate')
plt.title('Average Interest Rate vs. Grade')

plt.show()
```

```python


funded_amount = df['funded_amnt']
loan_amount = df['loan_amnt']
sns.barplot(x=loan_amount, y=funded_amount)

plt.xlabel('Loan Amount')
plt.ylabel('Funded Amount')
plt.title('Funded Amount vs. Loan Amount')

plt.show()
```

```python


emp_length = df['emp_length']
home_ownership = df['home_ownership']

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot emp_length
sns.countplot(x=emp_length, ax=axes[0])
axes[0].set_title('Distribution of Employment Length')
axes[0].set_xlabel('Employment Length')
axes[0].set_ylabel('Count')

# Plot home_ownership
sns.countplot(x=home_ownership, ax=axes[1])
axes[1].set_title('Distribution of Home Ownership')
axes[1].set_xlabel('Home Ownership')
axes[1].set_ylabel('Count')
plt.tight_layout()

plt.show()
```

```python



home_ownership = df['home_ownership']
annual_inc = df['annual_inc']

plt.figure(figsize=(10, 6))

sns.violinplot(x=home_ownership, y=annual_inc)

# Set labels and title
plt.xlabel('Home Ownership')
plt.ylabel('Annual Income (in millions)')
plt.title('Home Ownership vs. Annual Income')

plt.xticks(rotation=45)

# Format y-axis tick labels to display in millions
formatter = ticker.StrMethodFormatter('{x:.0f}M')
plt.gca().yaxis.set_major_formatter(formatter)


plt.tight_layout()
plt.show()
```

```python

loan_terms = df['term'].value_counts()
plt.style.use('default')

term_36 = loan_terms[' 36 months']
term_60 = loan_terms[' 60 months']


plt.bar(['36 months', '60 months'], [term_36, term_60], color=['#F266AB', '#A459D1'],label=['People with term 36 months','People with term 60 months'])

for i, count in enumerate([term_36,term_60]):
    plt.text(i, count,str(count), ha='center', va='bottom')
    
plt.title('Loan Term Distribution')
plt.xlabel('Loan Term')
plt.ylabel('Count')    
plt.legend()
plt.savefig('people count vs term plan.png',dpi=150)
plt.show()
```
